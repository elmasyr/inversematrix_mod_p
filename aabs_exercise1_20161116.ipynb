```python
''' import libraries and define some setup '''
import pandas as pd
%matplotlib inline
```

```python
df=pd.read_csv('group3.csv', header=0, sep=';', encoding='utf-8') # read data
df.shape # check size
```

```python
# get only the valid values and ignore additional white space
df=df[df.index<7458]
df.shape # dataset size
```

```python
# checks
print df.head()
print df.tail()
```

```python
# rename column
df.columns = ['ts', 'signal']
# make new variable "seconds"
df['index']=df.index
df['sec']=df['index']/20
```

```python
# checks
print df.head()
print df.tail()
```

## ploting the whole data
ticks on x-axis should show the different exercise steps

```python
df.plot(x=['sec'],y= ['signal'],xticks=[60, 120, 180, 240, 300, 330, 390], figsize=(12, 6));
```

### Task 1 with non-filtered data
15 sec from 1. Sitting still, both hands on the table, breathing normally 1 min 

```python
from numpy import *
import numpy as np

s = 30 # start time in seconds
e = 45 # end time in seconds

df_temp=df[(df.sec>=s) & (df.sec<e)]
data = np.array(df_temp.signal)
lmax = ((diff(sign(diff(data))) < 0)).nonzero()[0] + 1 # local max

print len(lmax), 'peaks in', (e-s), 'seconds'
df_temp.plot(x=['sec'],y= ['signal'], figsize=(12, 6));
```
